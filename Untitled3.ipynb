```python
from sklearn.linear_model import LinearRegression
import numpy as np

#Data ya Wiki ya Mwanza
#x=mawingu. 0=jua kali, 1=mawingu mengi
mawingu=np.array([[0],[1],[0],[0],[1],[0],[1]])
power=np.array([82,65,80,78,58,75,70])

print("Data imepangwa. Tukotayari kufundisha AI")
```

```python
# Tunaunda na kufundisha AI
ai=LinearRegression()
ai.fit(mawingu,power)

print("AI imeshafunzwa!")
```

```python
#1.UTABIRI WA MAWINGU
kesho_mawingu=np.array([[1]]) #kutakuwa na mawingu
utabiri=ai.predict(kesho_mawingu)

print("Utabiri wa Power ya Kesho:",round(utabiri[0],2),"W")
```

```python
#2.UTABIRI WA JUA
kesho_jua_kali=np.array([[0]]) # 0=kutakuwa na jua kali
utabiri=ai.predict(kesho_jua_kali)

print("Utabiri wa Power ya Kesho:",round(utabiri[0],2),"W")
```

```python
# #3.RIPOTI YA ENGINEER - %LOSS
#Tunatumia matokeo tuliyoyapata tayari hapo juu
power_jua=78.75 #kutoka Cell[4]
power_mawingu=64.33 #kutoaka Cell[3]

tofauti=power_jua-power_mawingu
percent_loss=(tofauti/power_jua)*100

print("\n---RIPOTI YA MATOKEO YA ENGINEER---")
print("Power wakati wa Jua:",power_jua,"W")
print("Power wakati wa Mawingu:",power_mawingu,"W")
print("Asilimia ya Loss:",round(percent_loss,2), "%")

```
